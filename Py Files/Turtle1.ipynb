```python
from mplturtle import *
```

```python
reset()

left(45)
forward(35)
right(90)
forward(35)
right(135)
forward(50)
```

```python

def square(size):
    for i in range(4):
        forward(size)
        right(90)
    

def triangle():
    left(45)
    forward(35)
    right(90)
    forward(35)
    right(135)
    forward(50)
        
        
```

```python
reset()
square(40)
```

```python
animate()
```

```python
def house():
    square(50)
    triangle()
    right(180)
    forward(50)

def moveRight():
    penup()
    forward(50)
    pendown()

```

```python
reset()
def fourhouses():
    house()
    moveRight()

    house()
    moveRight()

    house()
    moveRight()

    house()
    
def moveDown(): 
    penup()
    right(90)
    forward(100)
    right(90)
    forward(50)
    right(180)
    pendown()
    
def wayleft():
    penup()
    left(180)
    forward(350)
    right(180)
    pendown()
```

```python
animate()
```

```python
reset()
def neighborhood():
    fourhouses()
    moveDown()
    house()
    wayleft()
    house()
    moveDown()
    fourhouses()
```

```python
reset()
neighborhood()
```

```python
animate()
```
