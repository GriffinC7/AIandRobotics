```python
def square(size):
    for i in range(4):
        forward(size)
        right(90)

def square_with_color(size,color):

    if color=="red":
        pencolor("red")
    elif color=="green":
        pencolor("blue")
    else:
        pencolor("black")
        
    for i in range(4):
        forward(size)
        right(90)
        
def moveRight():
    penup()
    forward(20)
    pendown()
    
def newRow():   
    penup()
    right(90)
    forward(20)
    right(90)
    forward(100)
    right(180)
    pendown()
    
def firstRow():
    for i in range (2):
        square_with_color(15,'black')
        moveRight()
        square_with_color(15,'red')
        moveRight()
        
    square_with_color(15,'black')
    moveRight() 
    
def secondRow():
    for i in range (2):
        square_with_color(15,'red')
        moveRight()
        square_with_color(15,'black')
        moveRight() 
        
    square_with_color(15,'red')
    moveRight()
def blackredgrid():
    for i in range (2):
        firstRow()    
        newRow()
        secondRow()
        newRow()
    firstRow()
```

```python
reset()
blackredgrid()
```
